```python
#image to data

from pytesseract import Output
```

```python
img = cv2.imread('/home/abhay/Documents/DL Projects/ocr/image-242.png')
```

```python
cv2.imshow('image',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img_data = pytesseract.image_to_data(img,output_type=Output.DICT)
```

```python
img_data
```

```python
img_data.keys()  #printing the keys in a dictionary
```

```python
img_data['text']   #left top width height == x,y,w,h
```

```python
img_data['conf']
#negative value shows space
```

```python
n_boxes = len(img_data['text'])
n_boxes
```

```python
for i in range(n_boxes):
     if img_data['conf'][i] > 90:
        x,y,w,h = img_data['left'][i],img_data['top'][i],img_data['width'][i],img_data['height'][i]
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),thickness=3)

cv2.imshow('image',img)
cv2.waitKey()
cv2.destroyAllWindows()
print(x,y,w,h)

```
