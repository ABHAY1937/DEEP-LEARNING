```python
import tensorflow as tf
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/LoanApprovalPrediction.csv')
df
```

```python
df.shape
```

```python
df.dtypes
```

```python

#convert object to numerical values using label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df.Gender = le.fit_transform(df.Gender)
df.Married = le.fit_transform(df.Married)
df.Education = le.fit_transform(df.Education)
df.Self_Employed = le.fit_transform(df.Self_Employed)
df.Property_Area = le.fit_transform(df.Property_Area)
df.Loan_Status = le.fit_transform(df.Loan_Status)
```

```python
df
```

```python
df.isna().sum()
```

```python
first = df["LoanAmount"].mean()
df["LoanAmount"].fillna(first,inplace =True)
#
term = df["Loan_Amount_Term"].mean()
df["Loan_Amount_Term"].fillna(term,inplace = True)
#
history = df["Credit_History"].mean()
df["Credit_History"].fillna(history, inplace = True)
```

```python
df.drop(columns=["Dependents","Loan_ID"] ,inplace=True,axis=1)
```

```python
df.isna().sum()
```

```python
df
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
x.ndim
```

```python
y.ndim
```

```python
#traning model
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
#normalization
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
scalar.fit(x_train)
x_train = scalar.fit_transform(x_train)
x_test = scalar.fit_transform(x_test)
```

```python
#model creation
ann = tf.keras.models.Sequential()
```

```python
#1st hidden layer
ann.add(tf.keras.layers.Dense(7,activation='relu'))
#2nd hidden layers
ann.add(tf.keras.layers.Dense(5,activation="relu"))
#3rd hidden layers
ann.add(tf.keras.layers.Dense(6,activation='relu'))
```

```python
#output layer
ann.add(tf.keras.layers.Dense(1,activation='sigmoid'))
```

```python
#complilling ANN
ann.compile(optimizer='adam', loss='binary_crossentropy',metrics=["accuracy"])
```

```python
ann.fit(x_train,y_train,batch_size=32,epochs=200)
```

```python
pred = ann.predict(scalar.transform([[1,0,0,0,5849,0.0,145,300.0,1.0,1]]))
if(pred>0.5):
  print(1)
else:
  print(0)


```
